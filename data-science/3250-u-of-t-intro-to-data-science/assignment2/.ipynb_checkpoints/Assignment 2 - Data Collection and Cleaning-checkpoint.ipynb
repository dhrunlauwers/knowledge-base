# Code and analysis

## Importing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data-files/train.csv', index_col='PassengerId', na_values={'Fare':[0]})

df.info()
```

## Identify missing data
Train data set has 891 entries, and data is missing for some columns:
* Age only has values for 714 entries (80%)
* Cabin only has values for 204 entries (23%)
* Fare only has values for 876 entries (98%)
* Embarked only has 889 entries (99%)

```python
(df.count() / len(df)).sort_values()
```

## Examine distribution of numerical data

```python
df[['Age', 'Pclass', 'SibSp', 'Parch', 'Fare']].describe()
```

## Clean up data
* Missing Fare will be adjusted to the average fare for each class of passenger because fares in upper class are higher than fare in lower class.
* Missing Age will be adjusted to the average age for each class and sex of passenger because passengers in upper class are older than those in lower class, and men are older than women.
* Missing Cabin will be adjusted to reflect the most common value in the original dataset
* Missing Embarked will be adjusted to reflect the most common value in the original dataset

```python
df['Fare'] = df.groupby('Pclass').transform(lambda x: x.fillna(x.mean()))['Fare']
df['Age'] = df.groupby(['Pclass', 'Sex']).transform(lambda x: x.fillna(x.mean()))['Age']
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].value_counts().idxmax())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].value_counts().idxmax())
```

### Check to make sure clean-up is complete
1. No more missing values should be present
2. Numerical values should have minimum and maximum values that make sense

```python
df.count() / len(df) == 1
```

```python
df[['Age', 'Pclass', 'SibSp', 'Parch', 'Fare']].describe()
```

## Feature generation
* Passengers could be split into children and adults based on age
* Sex, Embarked and Adult or Child can be one-hot encoded to show correlation more clearly
* Total family members on board could be added up

```python
def split_by_age(age):
    if age > 14:
        return 'Adult'
    else:
        return 'Child'

df['Adult or Child'] = df['Age'].apply(split_by_age)
df = pd.concat([df, pd.get_dummies(df['Sex'])], axis=1)
df = pd.concat([df, pd.get_dummies(df['Adult or Child'])], axis = 1)
df = pd.concat([df, pd.get_dummies(df['Embarked'])], axis = 1)
df['Total Family Members'] = df['Parch'] + df ['SibSp']
df.head()
```

## Data analysis
1. Correlations:
    * Some correlations support hypothesis that females and children in first class were given priority:
        * Strong positive correlations exist with females, passengers in higher classes, and children
        * Strong nevative correlations exist with males, passengers in lower classes, and adults
    * Some correlations are likely due to other factors:
        * Fare, Embarkment: C and Embarkment: S are also correlated with Pclass
2. Survival rates: Correlations are reflected in the survival rate for each type of passenger. 
    * Overall, about 61% of passengers survived.
    * Surival rate depends on passenger class:
        * 63% of passengers in first class survived 
        * 47% of passengers in second class survived
        * 24% of passengers in thrid class surivived
    * Surival rate depends on sex:
        * 74% of female passengers survived
        * 19% of male passengers survived
    * Survival rate depends on age:
        * 58% of children surivived
        * 36% of adults survived

```python
#Overall survival rate
df['Survived'].value_counts() / len(df)
```

```python
# Survival rate by class
df.groupby('Pclass')['Survived'].agg(['count','sum', 'mean'])
```

```python
# Survival rate by sex
df.groupby('Sex')['Survived'].agg(['count','sum', 'mean'])
```

```python
# Adult vs Child survival rate
df.groupby('Adult or Child')['Survived'].agg(['count','sum', 'mean'])
```

```python
# Correlations
df.corr()['Survived'].sort_values()
```

```python
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(10, 220, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df.corr(), mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
