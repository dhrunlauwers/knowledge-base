```python
import pandas as pd
import matplotlib as plt
import datetime
```

```python
presidents = pd.read_csv('data-sets/us_presidents.csv')
```

```python
presidents['start'] = pd.to_datetime(presidents['start'])
presidents['end'] = pd.to_datetime(presidents['end'])
```

```python
def clean_party(party):
    if ((party == 'Republican') | (party == 'Democratic')):
        return party
    else:
        return 'Other'
```

```python
presidents['party'] = presidents['party'].apply(clean_party)
```

```python
presidents
```

```python
years = pd.date_range(presidents['start'].min(), datetime.date.today(), freq='Y')
```

```python
presidents_by_year = pd.DataFrame(years, columns=['Year'])
```

```python
presidents_by_year.info()
```

```python
def find_president(year):
    president = presidents[(year > presidents['start']) & (year < presidents['end'])]
    return president['president'].values
```

```python
def find_party(year):
    president = presidents[(year > presidents['start']) & (year < presidents['end'])]
    return president['party'].values
```

```python
presidents_by_year['president'] = presidents_by_year['Year'].apply(find_president)
```

```python
presidents_by_year['party'] = presidents_by_year['Year'].apply(find_party)
```
