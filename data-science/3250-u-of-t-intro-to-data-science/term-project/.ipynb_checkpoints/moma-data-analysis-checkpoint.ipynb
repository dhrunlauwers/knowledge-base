## Data cleanup

```python
def calculate_artist_age(dates):
    begin_date, end_date = dates
    if begin_date < 1:
        return None
    elif end_date < 1:
        return 2019 - begin_date
    else:
        return end_date - begin_date
```

```python
artists['Age'] = artists[['BeginDate','EndDate']].apply(calculate_artist_age, axis=1)
```

```python
artists['Gender'] = artists['Gender'].str.lower()
```

```python
artworks['YearAcquired'] = pd.to_numeric(artworks['DateAcquired'].str.split('-').str[0])
```

```python
exhibitions['ExhibitionBeginDate'] = pd.to_datetime(exhibitions['ExhibitionBeginDate'])
```

```python
exhibitions['ExhibitionEndDate'] = pd.to_datetime(exhibitions['ExhibitionEndDate'])
```

```python
exhibitions['Year'] = exhibitions['ExhibitionBeginDate'].dt.year
```

### What nationality do most artists have?

```python
top_artist_nations = artists['Nationality'].value_counts()[1:11] / artists['Nationality'].value_counts().sum()
```

```python
artists['Nationality'].value_counts()[1].sum() /  artists['Nationality'].value_counts().sum()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_ylabel('Proportion of artists')
ax.set_xlabel('Country of origin')
ax.set_title('Artists by country of origin')
sns.barplot(top_artist_nations.index, top_artist_nations.values)
plt.savefig('3-nationality.png')
```

### Which artists have the most works in the collection?

```python
top_artists = artworks['Artist'].value_counts()[:10]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_ylabel('Number of artworks')
ax.set_xlabel('Artist name')
ax.set_title('Number of artworks by artist')
plt.xticks(rotation=30)
sns.barplot(top_artists.index, top_artists.values)
plt.savefig('4-artists.png')
```

### How old are the artists?

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Age distribution of artists')
sns.distplot(artists['Age'], bins=100)
plt.savefig('5-artists-age.png')
```

### What gender are the artists?

```python
artists['Gender'].value_counts()[:2]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_title('Age distribution of artists')
ax.pie(artists['Gender'].value_counts()[:2], labels=['male', 'female'], autopct='%1.1f%%')
plt.savefig('6-artists-sex.png')
```

```python
plt.rc('font', size=10)
```

### Which mediums are most popular?

```python
top_mediums = artworks['Medium'].value_counts()[:10]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_ylabel('Number of artworks')
ax.set_xlabel('Medium name')
ax.set_title('Number of artworks by medium')
plt.xticks(rotation=30)
sns.barplot(top_mediums.index, top_mediums.values)
plt.savefig('10-medium.png')
```

### How many works are added each year?

```python
artworks_by_year = artworks[artworks['YearAcquired'] >= 1929]['YearAcquired'].value_counts()
artworks_by_year.sort_index(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(16,9))

#add title, axis labels, legend
ax.set_title('Artworks purchased by MoMa')
ax.set_ylabel('Number of Artworks Purchased')
ax.set_xlabel('Time')

artworks_by_year.plot(color='black')
plt.savefig('12-artworks-added.png')
```

### How many exhibitions have there been? How many each year?

```python
len(exhibitions['ExhibitionID'].unique())
```

```python
exhibits_by_year = exhibitions.groupby(['Year'])['ExhibitionID'].nunique()
```

```python
exhibits_by_year.describe()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

#add title, axis labels, legend
ax.set_title('Exhibitions held by by MoMa')
ax.set_ylabel('Number of Exhibitions held')
ax.set_xlabel('Time')

exhibits_by_year.plot(color='black')
plt.savefig('13-exhibits.png')
```

### Which artist is featured in the most exibitions?

```python
top_exhibitors = exhibitions['DisplayName'].value_counts()[:10]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_ylabel('Number of exhibits')
ax.set_xlabel('Artist name')
ax.set_title('Number of exhibits by artist')
plt.xticks(rotation=45)
sns.barplot(top_exhibitors.index, top_exhibitors.values)
plt.savefig('7-top-exhibitors.png')
```

The chart above actuall includes some Curators, not just Artists..

```python
top_artists = exhibitions['DisplayName'].where(exhibitions['ExhibitionRole'] == 'Artist').value_counts()[:20]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.set_ylabel('Number of exhibits')
ax.set_xlabel('Artist name')
ax.set_title('Number of exhibits by artist')
plt.xticks(rotation=45)
sns.barplot(top_artists.index, top_artists.values)
plt.savefig('7-top-exhibitors.png')
```
