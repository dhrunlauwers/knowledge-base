**Task 1:** Read the data and create a time series.
- HINT: you want to look at the parameters of read_csv() (Links to an external site.) function to make sure that you create a Series and not a Dataframe and that the dates are formatted as datetime object and used as an index.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot
%matplotlib inline
```

```python
beer = pd.read_csv('data\\monthly-beer-production-in-austr.csv', index_col=['Month'], squeeze=True, parse_dates=['Month'])
```

```python
type(beer)
```

**Task 2:** Plot the time series. What do you see in the graph? Does the data have seasonality or trend? Describe the graph. Is it stationary?

```python
beer.plot()
```

>The data has seasonality because of frequent recurring peaks and valleys, and the data is trending towards more overall beer production as time gets closer to the end of the series. Because of this, the graph is not stationary.

**Task 3:** Plot an autocorrelation function graph of this time series. Is data autocorrelated or not? 

```python
autocorrelation_plot(beer)
```

> The data is autocorrelated because it shows several lag values with significant correlation.

**Task 4:** Perform differencing on the time series - calculate the difference of each element of the Series compared with another element in the Series (you can use default setting of the function which calculates the difference with the element in the previous row). You can use either function diff() (Links to an external site.), or shift() which we looked at in the class.

```python
beer12 = beer.diff(periods=12)
```

**Task 5:** Plot a time plot and an autocorrelation function graph of this new time series. Is data autocorrelated or not? Is it stationary?

*HINT:* Autocorrelation function will not display any data if Series contain any missing values. Make sure that you deal with NaNs before plotting.

```python
beer12 = beer12.dropna()
```

```python
beer12.plot()
```

```python
autocorrelation_plot(beer12)
```

> The time series after differencing is more stationary as the vast majority of lag values do not have significant correlation. That being said, since some lag values have significant correlation, the time series cannot be considered fully stationary.


**Task 6:** Simple validation of stationarity for the differenced time series. Split the time series (from task 4) into 2 subsets and calculate mean and variance for each subset. Can we make any conclusion about stationarity of the time series based on this calculation?

```python
split = int(len(beer12) / 2)
beer_a = beer12[:split]
beer_b = beer12[split:]
```

```python
beer_a.mean()
```

```python
beer_b.mean()
```

```python
beer_a.var()
```

```python
beer_b.var()
```

> Stationary time series should have the same mean and variance for the entire time series. After splitting the time series, it is clear that the mean and variance are different, so the time series should not be considered stationary.

**Task 7 (optional, 5 bonus points):** Using the Augmented Dickey–Fuller test which we discussed in the class, calculate ADF Statistic value, p-value and critical values for the original time series and time series you get in Task 4. Describe the result. 

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf = adfuller(beer)
print('ADF Statistic: %f' % adf[0])
print('p-value: %f' % adf[1])
print('Critical Values:')
for key, value in adf[4].items():
    print('\t%s: %.3f' % (key, value))
```

> Test statistic = -2.28 which is greater than -3.445 at 1% confidence and at any other confidence level. This means we cannot reject the hypothesis, and the TS are not stationary and has a trend.

> Also, p-value is > 0.05 which tells us we cannot reject the hypothesis that the TS is not stationary.
