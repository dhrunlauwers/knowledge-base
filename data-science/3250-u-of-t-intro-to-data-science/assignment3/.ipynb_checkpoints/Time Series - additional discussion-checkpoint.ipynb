### Time Series revisited

```python
import pandas as pd
import matplotlib as plt
#plt.style.use(style='ggplot')
plt.rc('figure', figsize=(40, 30))
%matplotlib inline
```

```python
elect = pd.read_csv('Electric_Production.csv', index_col=['DATE'], parse_dates=['DATE'], squeeze=True)
elect.head()
```

```python
elect.plot(figsize=(10, 7))
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(elect)
```

The Augmented Dickey–Fuller test is checking for stationarity or non-stationary data.  The test is trying to reject the null hypothesis that a unit root exists and the data is non-stationary. If the null hypothesis is rejected, then the alternate can be considered valid (e.g., the data is stationary). 

Hence the **Null Hypothesis** is: The data is non-stationary and has some time dependent structure.

When we run the test, we get an **ADF Statistic** value and a **p-value**. The ADF number should be a negative number. The more negative it is, the stronger the rejection of the Null Hypothesis. The p-value should be beneath a certain threshold value (e.g., 1% or 5%, etc) for a confidence level. 

We will use use 5% (or 95% confidence level). This means:

If the p-value is greater than 0.05 then we say we cannot reject the null hypothesis because the data has a unit root and is **non-stationary**.  

If the p-value is less than or equal to 0.05, we can say we reject the null hypothesis because the data does not have a unit root and is stationary.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adf_electr = adfuller(elect)
print('ADF Statistic: %f' % adf_electr[0])
print('p-value: %f' % adf_electr[1])
print('Critical Values:')
for key, value in adf_electr[4].items():
    print('\t%s: %.3f' % (key, value))
```

Test statistic = -2.26 which is greater than -3.448 at 1% confidence and at any other confidence level. This means we cannot reject the hypothesis, and the TS are not stationary and has a trend.

Also, p-value is > 0.05 which tells us we cannot reject the hypothesis that the TS is not stationary.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
ts_decompose = seasonal_decompose(elect, model='additive')
```

```python
ts_decompose.plot();
```

```python
autocorrelation_plot(ts_decompose.resid.dropna())
```

```python
adf_electr_resid = adfuller(ts_decompose.resid.dropna())
print('ADF Statistic: %f' % adf_electr_resid[0])
print('p-value: %f' % adf_electr_resid[1])
print('Critical Values:')
for key, value in adf_electr_resid[4].items():
    print('\t%s: %.3f' % (key, value))
```
